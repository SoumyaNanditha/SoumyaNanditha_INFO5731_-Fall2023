```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Define the URL to scrape
url = 'https://digitaldreamdoor.com/pages/best_songs_1000_popular_1-500.html'

# Send an HTTP GET request to the URL and get the response
response = requests.get(url)

# Parse the HTML content of the page using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the song data div elements on the page
song_data = soup.find_all('div', class_='song')

# Initialize an empty list to store song information
song_info = []

# Loop through each song data div element and extract relevant information
for song in song_data:
    # Extract song title
    title = song.find('h3').text.strip()

    # Extract artist name
    artist = song.find('p').text.strip()

    # Extract song ranking
    ranking = song.find('span', class_='songnum').text.strip()

    # Append the extracted information to the song_info list as a dictionary
    song_info.append({
        'Song Title': title,
        'Artist': artist,
        'Ranking': ranking
    })

# Create a Pandas DataFrame from the collected song information
song_df = pd.DataFrame(song_info)
song_df.head()

```

Question 3 (10 points): Write python code to collect 1000 articles from Google Scholar (https://scholar.google.com/), Microsoft Academic (https://academic.microsoft.com/home), or CiteSeerX (https://citeseerx.ist.psu.edu/index), or Semantic Scholar (https://www.semanticscholar.org/), or ACM Digital Libraries (https://dl.acm.org/) with the keyword "information retrieval". The articles should be published in the last 10 years (2013-2023).

The following information of the article needs to be collected:

(1) Title

(2) Venue/journal/conference being published

(3) Year

(4) Authors

(5) Abstract

```python
# Import necessary libraries
import requests
from bs4 import BeautifulSoup

# Define a function for web scraping scholarly articles
def webscrape_schol_articles(url):
    # Send an HTTP GET request to the provided URL
    response = requests.get(url)
    
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find all the div elements with class 'gs_ri' which represent individual articles
    articles = soup.find_all('div', class_='gs_ri')

    # Loop through each found article
    for article in articles:
        
        # Extract the title of the article from the 'h3' tag with class 'gs_rt'
        title = article.find('h3', class_='gs_rt').text
        
        # Extract the publication venue, year, and authors from the 'div' tag with class 'gs_a'
        venue = article.find('div', class_='gs_a').text.split(' - ')[-1]
        year = article.find('div', class_='gs_a').text.split(' - ')[0]
        authors = article.find('div', class_='gs_a').text.split(' - ')[1]
        
        # Extract the abstract of the article from the 'div' tag with class 'gs_rs'
        abstract = article.find('div', class_='gs_rs').text

        # Print the extracted information for the article
        print("Title:", title)
        print("Venue:", venue)
        print("Year:", year)
        print("Authors:", authors)
        print("Abstract:", abstract)
        print("\n---\n")

# Loop through plenty of pages of results to get 1,000 articles.
articles_per_page = 10
total_articles = 1000

for page in range(0, total_articles, articles_per_page):
    # Construct the URL for the Google Scholar search results page
    page_url = f"https://scholar.google.com/scholar?hl=en&as_sdt=0%2C5&q=information+retrieval&as_ylo=2013&as_yhi=2023&start={page}"
    # Call the webscrape_schol_articles function to scrape and print articles from the current page
    webscrape_schol_articles(page_url)
```
